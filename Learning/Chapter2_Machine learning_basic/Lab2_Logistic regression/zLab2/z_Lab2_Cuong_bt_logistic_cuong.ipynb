```python
# Thêm thư viện
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Hàm sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

# Check Input

```python
# Load data từ file csv
data = pd.read_csv('data_logistic.csv').values

N, d = data.shape
x = data[:, 0:d-1].reshape(-1, d-1)
y = data[:, 2].reshape(-1, 1)

print(data[:5])
```

# Data Visualization

```python
# Vẽ data bằng scatter
import matplotlib.pyplot as plt

# Tách dữ liệu thành hai danh sách riêng biệt cho các lớp
class_0 = [x[i] for i in range(len(x)) if y[i][0] == 0]
class_1 = [x[i] for i in range(len(x)) if y[i][0] == 1]

# Trực quan hóa dữ liệu
plt.scatter([point[0] for point in class_0], [point[1] for point in class_0], color='red', label='Class 0')
plt.scatter([point[0] for point in class_1], [point[1] for point in class_1], color='blue', label='Class 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Visualizing X and y')
plt.legend()
plt.show()
```

# Training by lib 

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression

# Khai báo mô hình Logistic Regression
logreg = LogisticRegression()

X_train = x[:70]
X_test = x[70:]
y_train = y[:70]
y_test = y[70:]

# Huấn luyện mô hình
logreg.fit(X_train, y_train.ravel())  # Sử dụng ravel() để biến đổi thành mảng 1D

# Dự đoán trên tập test
y_pred = logreg.predict(X_test)

# In các giá trị dự đoán và giá trị thực tế từ tập test
print("Predicted labels:", y_pred)
print("True labels:", *y_test.reshape(1, -1))

# Tính toán độ chính xác
accuracy = accuracy_score(y_test.ravel(), y_pred)
print("Accuracy:", accuracy)
```

# Trainning by NO lib

```python
import warnings

# Tắt các cảnh báo RuntimeWarning
warnings.filterwarnings("ignore", category=RuntimeWarning)

# Thêm cột 1 vào dữ liệu x
x = np.hstack((np.ones((N, 1)), x))
w = np.array([0.,0.1,0.1]).reshape(-1,1)


from sklearn.model_selection import train_test_split
# Chia dữ liệu thành tập train và test
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)



# Số lần lặp bước 2
numOfIteration = 1000
cost = np.zeros((numOfIteration,1))
learning_rate = 0.01
for i in range(1, numOfIteration):
    # Tính giá trị dự đoán
    y_predict = sigmoid(np.dot(X_train, w))
    cost[i] = -np.sum(np.multiply(y_train, np.log(y_predict)) + \
    np.multiply(1-y_train, np.log(1-y_predict)))

    # Gradient descent
    w = w - learning_rate * np.dot(X_train.T, y_predict-y_train)
    # print(cost[i])

# Lưu weight dùng numpy.save(), định dạng '.npy'
np.save('weight_logistic.npy', w)
```

# Plot Loss Function

```python
# Vẽ biểu đồ hàm loss function
plt.plot(cost[1:])
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Loss Function over Iterations')
plt.show()
```

# Predict


```python
# Load weight từ file '.npy'
w = np.load('weight_logistic.npy')
print(w)
y_predict_test = sigmoid(np.dot(X_test, w))
# print(y_predict_test)
y_result = [0 if i<0.5 else 1 for i in y_predict_test]
print(y_result[:5])
print(y_test[:5])
```

# Accuracy

```python
tp = 0
for i in range(len(y_result)):
    if (y_result[i] == y_test[i]):
             tp += 1

print("Độ chính xác thu cong:", tp/len(y_result))

from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, y_result)
print('Recall thu vien: ', score)
```

# Recall

```python
tp = 0
fn = 0
for i in range(len(y_result)):
    if (y_test[i] == 1):
        if (y_result[i] == y_test[i]):
            tp += 1
        else:
            fn += 1

recall = tp/(tp+fn)
print("Recall thu cong:", recall)


from sklearn.metrics import recall_score
score = recall_score(y_test, y_result, average='binary')
print('Recall thu vien: ', score)
```

# F1-Score

```python
tp = 0
fp = 0
for i in range(len(y_result)):
    if (y_test[i] == 1):
        if (y_result[i] == y_test[i]):
            tp += 1
    else:
        if (y_result[i] != y_test[i]):
            fp += 1

precision = tp/(tp+fp)

f1 = 2*precision*recall / (precision + recall) 
print("F1-score thu cong: ", f1)


from sklearn.metrics import f1_score
score = f1_score(y_test, y_result, average='binary')
print('F1-score thu vien: ',score)
```
