```python
import pandas as pd
import numpy as np

def my_linear_regression(w_init, X, y, lr, esilon, epoches=10000):
    # epoches = số lần lặp , lr = tốc độ học , esilon
    w = w_old = w_init
    ep = 0
    N, d = X.shape[0], X.shape[1]
    print(N, d) # 4000 , 6 
    # Vòng lặp Gradient Descent:
    # Lặp lại cho đến khi tiêu chuẩn hội tụ (epsilon) được đáp ứng hoặc số lượng vòng lặp tối đa (epoches) được đạt.
    while ep < epoches:
        w = w - lr*np.dot(X.T, (np.dot(X, w) - y))/N # update w 

        # Kiểm tra điều kiện dừng: Nếu độ lớn của đạo hàm của hàm mất mát (tính theo trung bình) nhỏ hơn giá trị epsilon 
        # thì thoát vòng lặp chứ không cần đợi đến lúc epoches đủ 10000
        if np.linalg.norm(w - w_old)/d < esilon: 
            print(np.linalg.norm(w - w_old)/d)
            break

        w_old = w

        ep += 1
        # cứ mỗi 1000 lần lặp thì in ra xem 
        if ep % 1000 == 0:
            print("epoches = ", ep, end=" ")
            print("loss = ", np.linalg.norm(np.dot(X, w) - y)/N)
        
    return w


if __name__ == "__main__":
    X = []
    # Code you have previously used to load data
    iowa_file_path = './USA_Housing.csv'
    data = pd.read_csv(iowa_file_path)
    df = pd.DataFrame(data)
    # xử lý dữ liệu
    data = df.dropna()
    data = data.drop(["Address"], axis=1) 

    # Optimize code 
    def preprocess_data(data):
        # Loại bỏ các dòng có giá trị 0 trong các cột quan trọng
        cols_to_check = ["Avg. Area Income", "Avg. Area House Age", "Avg. Area Number of Rooms", "Avg. Area Number of Bedrooms", "Area Population", "Price"]
        data = data[(data[cols_to_check] != 0).all(axis=1)]

        # Min-max hóa dữ liệu
        for col in cols_to_check:
            data[col] = (data[col] - data[col].min()) / (data[col].max() - data[col].min())

        return data

    # Sử dụng hàm preprocess_data để tiền xử lý dữ liệu
    data = preprocess_data(data)
    # print(data)

    # Lấy các giá trị của các đặc trưng
    x1 = np.array(data["Avg. Area Income"]).reshape(-1, 1)
    x2 = np.array(data["Avg. Area House Age"]).reshape(-1, 1)
    x3 = np.array(data["Avg. Area Number of Rooms"]).reshape(-1, 1)
    x4 = np.array(data["Avg. Area Number of Bedrooms"]).reshape(-1, 1)
    x5 = np.array(data["Area Population"]).reshape(-1, 1)
    # print(x1.shape) # mang trận 5000x1 (5000 hàng và 1 cột)

    # data = data.drop(data[data["Avg. Area Income"] == 0].index)
    # data = data.drop(data[data["Avg. Area House Age"] == 0].index)
    # data = data.drop(data[data["Avg. Area Number of Rooms"] == 0].index)
    # data = data.drop(data[data["Avg. Area Number of Bedrooms"] == 0].index)
    # data = data.drop(data[data["Area Population"] == 0].index)
    # data = data.drop(data[data["Price"] == 0].index)

    # # min và max hóa dữ liệu
    # data["Avg. Area Income"] = (data["Avg. Area Income"] - data["Avg. Area Income"].min()) / (data["Avg. Area Income"].max() - data["Avg. Area Income"].min())
    # data["Avg. Area House Age"] = (data["Avg. Area House Age"] - data["Avg. Area House Age"].min())/(data["Avg. Area House Age"].max()-data["Avg. Area House Age"].min())
    # data["Avg. Area Number of Rooms"] = (data["Avg. Area Number of Rooms"] - data["Avg. Area Number of Rooms"].min())/(data["Avg. Area Number of Rooms"].max()-data["Avg. Area Number of Rooms"].min())
    # data["Avg. Area Number of Bedrooms"] = (data["Avg. Area Number of Bedrooms"] - data["Avg. Area Number of Bedrooms"].min())/(data["Avg. Area Number of Bedrooms"].max()-data["Avg. Area Number of Bedrooms"].min())
    # data["Area Population"] = (data["Area Population"] - data["Area Population"].min())/(data["Area Population"].max()-data["Area Population"].min())
    # data["Price"] = (data["Price"] - data["Price"].min())/(data["Price"].max()-data["Price"].min())
    
    # x1 = np.array(data["Avg. Area Income"]).reshape(-1, 1)
    # x2 = np.array(data["Avg. Area House Age"]).reshape(-1, 1)
    # x3 = np.array(data["Avg. Area Number of Rooms"]).reshape(-1, 1)
    # x4 = np.array(data["Avg. Area Number of Bedrooms"]).reshape(-1, 1)
    # x5 = np.array(data["Area Population"]).reshape(-1, 1)

    y = np.array(data["Price"]).reshape(-1, 1) # Một cột và n hàng , ma trận N*1 
    X = np.concatenate((x1, x2, x3, x4, x5), axis=1)
    # print(X)
    # print(X.shape) # 5000x5 (5000 hàng va 5 cột)
    # print(y.shape) # 5000x5 (5000 hàng va 5 cột)
    print("len(X) = ", len(X))
    X_test = X[4000:, :]
    y_test = y[4000:]
    X = X[:4000, :]
    y = y[:4000]
    # print("X=",X)
    # print("y=",y)
    X_bar = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
    # print(X)
    # print(X_bar)
    X_test_bar = np.concatenate((np.ones((X_test.shape[0], 1)), X_test), axis=1)
    A = np.dot(X_bar.T, X_bar) # nhân 2 ma trận 
    B = np.dot(X_bar.T, y)
    w1 = np.dot(np.linalg.pinv(A), B) # dựa theo công thức phía trên w = (X^T.X)^-1 . X^T . y # np.linalg.pinv chính là ^1 
    w_init = np.random.randn(X_bar.shape[1], 1)
    # 1e-10 = 1*(10^-10) = 0.0000000001
    # a * (b ** c) = a * b^c 
    w = my_linear_regression(w_init, X_bar, y, lr = 0.05, esilon = 1e-10, epoches = 100000)

    print("w = ", w)
    print("w1 = ", w1)
    result = np.concatenate((y.reshape(-1, 1), np.dot(X_bar, w).reshape(-1, 1), np.dot(X_bar, w1).reshape(-1, 1)), axis=1)
    mean_y = np.mean(y)
    # mae test
    print("len(X_test) = ", len(X_test))
    mae_test = np.mean(abs(np.dot(X_test_bar, w) - y_test))
    mae1_test = np.mean(abs(np.dot(X_test_bar, w1) - y_test))
    print("mae_test = ", mae_test)
    print("mae1_test = ", mae1_test)
    print("%mae_test/mean_y = ", mae_test/mean_y*100, "%")
    print("%mae1_test/mean_y = ", mae1_test/mean_y*100, "%")
    
    # mse test
    mse_test = np.mean((np.dot(X_test_bar, w) - y_test)**2)
    mse1_test = np.mean((np.dot(X_test_bar, w1) - y_test)**2)
    print("mse_test = ", mse_test)
    print("mse1_test = ", mse1_test)
```
