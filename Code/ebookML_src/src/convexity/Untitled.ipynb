```python
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np 

choose = 4

fig = plt.figure()
ax = fig.gca(projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)

X = np.arange(-6, 6, 0.025)
Y = np.arange(-6, 6, 0.025)
X, Y = np.meshgrid(X, Y)
if choose == 1:
	Z = X**2 + Y**2
elif choose == 2: 
	Z = np.abs(X) + np.abs(Y) 
elif choose == 3: 
	Z = X**2 - Y**2
elif choose == 4:
	Z = 0.1*(X**2 + Y**2 - 10*np.sin(np.sqrt(X**2 + Y**2)))

# ax.plot_surface(X, Y, Z, rstride=8, cstride=8, cmap=cm.jet)
ax.plot_surface(X, Y, Z, cmap = 'jet')
cset = ax.contour(X, Y, Z, zdir='z', offset=-5, cmap='jet')
# cset = ax.contour(X, Y, Z, zdir='x', offset=0, cmap=cm.coolwarm)
# cset = ax.contour(X, Y, Z, zdir='y', offset=0, cmap=cm.coolwarm)

ax.set_xlabel('$x$', fontsize = 15)
ax.set_xlim(-6, 6)
ax.set_ylabel('$y$', fontsize =15)
ax.set_ylim(-6, 6)
if choose == 1:
	ax.set_zlabel('$f(x, y) = x^2 + y^2$', fontsize = 14)
elif choose == 2:
	ax.set_zlabel('$f(x, y) = |x| + |y|$', fontsize = 14)
elif choose == 3:
	ax.set_zlabel('$f(x, y) = x^2 - y^2$', fontsize = 14)
elif choose == 4:
	# ax.set_zlabel('$f(x, y) = \\frac{1}{10}(x^2 + 2y^2 - 2\sin(xy)$', fontsize = 20)
	ax.set_zlabel('$f(x, y)$', fontsize = 14)
ax.set_title('$f(x, y) = \\frac{1}{10}(x^2 + y^2 - 10\sin(\sqrt{x^2 + y^2})$ (quasiconvex)', fontsize = 14)
ax.set_zlim(-5, 6)

from matplotlib.backends.backend_pdf import PdfPages
filename = 'quasiconvex.pdf'
with PdfPages(filename) as pdf:
    pdf.savefig(bbox_inches='tight')
plt.show()
```
