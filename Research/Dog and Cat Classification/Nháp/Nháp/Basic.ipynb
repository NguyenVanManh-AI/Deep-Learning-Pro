```python
def rgb_to_gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])

# Chuyển đổi thành ảnh xám
gray_img = rgb_to_gray(img)

# Hiển thị ảnh xám
plt.imshow(gray_img, cmap=plt.get_cmap('gray'))
plt.axis('off')  # Tắt trục
plt.show()
```
