```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

# Ví dụ dữ liệu
sentences = ["Đây là một đoạn văn", "Đoạn văn khác", "Câu khác nữa"]
labels = [0, 1, 1]  # Nhãn phân loại

# Tokenization
tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(sentences)
sequences = tokenizer.texts_to_sequences(sentences)
print(sequences)

# Padding
padded_sequences = pad_sequences(sequences, maxlen=100)
```

```python
padded_sequences
```

```python
# In từ điển từ (word index)
word_index = tokenizer.word_index
print(word_index)
```

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense

model = Sequential()
model.add(Embedding(input_dim=10000, output_dim=128, input_length=100))
model.add(LSTM(128)) # 128 node ẩn , output_dim chính là embeding 128
model.add(Dense(1, activation='sigmoid'))  # Cho bài toán phân loại nhị phân

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
import numpy as np

padded_sequences = np.array(padded_sequences)
labels = np.array(labels)

model.fit(padded_sequences, labels, epochs=10, batch_size=32, validation_split=0.2)
```

```python
# Ví dụ đánh giá mô hình
loss, accuracy = model.evaluate(padded_sequences, labels)
print(f"Loss: {loss}, Accuracy: {accuracy}")
```
